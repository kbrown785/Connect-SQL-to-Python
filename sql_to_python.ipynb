```python
#Connect to Oracle
import cx_Oracle

def getOraConnection():
    oradsn = cx_Oracle.makedsn("db28-1v1.econ.census.gov",
                               "1640",
                               service_name = "cfsnprd")
    oraDBConn = cx_Oracle.connect(user = "JBID",
                                  password = "password",
                                  dsn = oradsn)
    
    print("DB Version = ", oraDBConn.version)
    
    #Query: example query SELECTING user and sysdate FROM TABLE.
    query = "SELECT user, sysdate FROM DUAL"
    cursor = oraDBConn.cursor()
    result = cursor.execute(query).fetchall()
    print(str(result))

if __name__ == '__main__':
	getOraConnection()
```
